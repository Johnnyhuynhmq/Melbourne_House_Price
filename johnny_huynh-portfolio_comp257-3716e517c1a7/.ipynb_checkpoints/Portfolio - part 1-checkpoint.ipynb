```python
# Read the gpx file into a dataframe
import pandas as pd
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
from geopy.distance import vincenty
%matplotlib inline


tree = ET.parse('race1.gpx')
root = tree.getroot()

# define a namespace dictionary to make element names simpler
# this mirrors the namespace definitions in the XML files
ns ={'gpx': 'http://www.topografix.com/GPX/1/1', 'gpxtpx': 'http://www.garmin.com/xmlschemas/TrackPointExtension/v1'}
```

```python
# when we look for elements, we need to use the namespace prefix
trk = root.find('gpx:trk', ns)
trkseg = trk.find('gpx:trkseg', ns)
```

```python
# iterate all trkpt elements - the children of trkseg
data = []
for trkpt in trkseg:
    variables = []
    #get some properties from the attributesd
    lon= trkpt.attrib['lon']
    lat = trkpt.attrib['lat']
    # get values from the child elements
    ele = trkpt.find('gpx:ele', ns).text
    time = trkpt.find('gpx:time', ns).text
    # now dive into the extensions
    ext = trkpt.find('gpx:extensions', ns)
    power = ext.find('gpx:power', ns).text 
    tpext = ext.find('gpxtpx:TrackPointExtension', ns)
    temp = tpext.find('gpxtpx:atemp', ns).text
    hr = tpext.find('gpxtpx:hr', ns).text
    cad = tpext.find('gpxtpx:cad', ns).text
    variables = [lon, lat, ele, time, power, temp, hr, cad]
    data.append(variables)
    
# read data into a dataframe and name the variables
columns = ['Longtitude', 'Latitude', 'Elevation', 'Time', 'Power', 'Temperature', 'HeartRate', 'Cadence']
dataset = pd.DataFrame(data, columns = columns)

# convert time into date-time type and other variables into float type
dataset['Time'] = pd.to_datetime(dataset['Time'])
for i in ['Longtitude', 'Latitude', 'Elevation', 'Power', 'Temperature', 'HeartRate', 'Cadence']:
    dataset[i] = dataset[i].astype(float)
dataset.tail()
```

```python
# duration in seconds
total_time = dataset['Time'][len(dataset) - 1] - dataset['Time'][0]
total_time_in_seconds = total_time.total_seconds()

# total distance in meters using vincenty method
distance_in_meters = []
for i in range(0, len(dataset) - 1):
    starting_point = (dataset['Longtitude'][i], dataset['Latitude'][i])
    ending_point = (dataset['Longtitude'][i+1], dataset['Latitude'][i+1])
    distance = vincenty(starting_point, ending_point).meters
    distance_in_meters.append(distance)
total_distance = sum(distance_in_meters)
print('Total distance of rider 1 is', round(total_distance, 2), 'm')

# Average speed of the rider in m/s
average_speed_ms = round(total_distance / total_time_in_seconds, 2)
print("The average speed of the rider is", average_speed_ms, "m/s")

# Average speed of the rider in m/s
average_speed_kmh = round(average_speed_ms * 18/5, 2)
print("The average speed of the rider is",average_speed_kmh, "km/h")
```

As we can easily calculate the average speed of rider 1 based on total travelled distance and total time of him, his average speed is at 9.82 m/s or 35.35 km/h

```python
# histogram for the distribution of heart rate
fig = plt.figure(figsize=(10,10))
plt.hist(dataset['HeartRate'], bins=20, histtype ='bar', color='g')
plt.title('Histogram of Heart Rate')
plt.xlabel('bps')
plt.show()
```

The histogram shows that the highest frequent heart rate is around 140 to 160, especially at around 150. This number is normally expected for a person who is cycling. There are very unpopular points that heart rate increased to over 180, which is possibly because the rider gained elevation at that time.

```python
# histogram for the distribution of power
fig = plt.figure(figsize=(10,10))
plt.hist(dataset['Power'], bins=20, histtype ='bar', color='r')
plt.title('Histogram of Power')
plt.show()
```

The generated power is at 200 to 300 in average. However, there was a high frequency of very low amount of power generated during the cycle. This is also expected as in riding a bicycle, the athlete is not always pushing the pedals, especially from high elevation to low elevation when the bike can move itself. 

```python
# Because the time is tracked on a per-second basis, the distance travelled (m) each second is also the 
# speed calculated in m/s
speed_in_ms = distance_in_meters
# adding one more value when the rider finally stops
speed_in_ms.append(0)
# histogram for the distribution of speed
fig = plt.figure(figsize=(10,10))
plt.hist(speed_in_ms, bins=30, histtype ='bar', color='b')
plt.title('Histogram of Speed')
plt.xlabel('m / s')
plt.show()

```

The speed in average is 9 to 11 m/s, showed by the histogram. This result is supported by our calculation above. 

```python
# Power vs Speed scatterplot
fig = plt.figure(figsize=(10,10))
plt.scatter(dataset['Power'], speed_in_ms)
plt.title('Power vs Speed Scatterplot')
plt.xlabel('Power')
plt.ylabel('Speed (m/s)')
plt.show()
```

There seems not to be a clear relationship between power generated and speed. At any speed, the power generated can also be any amount. 

```python
# Power vs Heart Rate scatterplot
fig = plt.figure(figsize=(10,10))
plt.scatter(dataset['Power'], dataset['HeartRate'])
plt.title('Power vs Heart Rate Scatterplot')
plt.xlabel('Power')
plt.ylabel('Heart Rate (bps)')
plt.show()
```

It seems that low power correlating with low heart rate. However, the degree of correlation is not high. We cannot conclude anything in this case.

```python
# Quick visualization of the map the rider passed through
fig = plt.figure(figsize=(10, 10))
plt.scatter(dataset['Longtitude'], dataset['Latitude'])
plt.title('Map')
plt.show()
```

```python
# Counting the number of Laps
count = 0
for i in range(1, len(dataset) - 1):
    starting_point = (dataset['Longtitude'][0], dataset['Latitude'][0])
    similar_point = (dataset['Longtitude'][i], dataset['Latitude'][i])
    distance = vincenty(starting_point, similar_point).meters
    if distance < 10:
        count += 1
print(count)
```

A scatterplot of GPS coordinates proves that the rider did cycle following a fixed lap. There are several ways to calculate the number of laps completed. The code above is one of the example. However, we are not sure how far the 'distance' variable needs to be less than to be counted as one lap, as in big road, the distance can vary. Therefore, I decided to use visualization to estimate the total distance of each lap. 
Firstly, let try plotting the first 300 data points.


```python
# Map for first 300 data points
fig = plt.figure(figsize=(10, 10))
plt.scatter(dataset['Longtitude'][0:300], dataset['Latitude'][0:300])
plt.title('Map')
plt.show()
```

As we can see, the rider has not completed one lap. Let do further 70 data points, to 370


```python
# Map for first 370 data points
fig = plt.figure(figsize=(10, 10))
plt.scatter(dataset['Longtitude'][0:370], dataset['Latitude'][0:370])
plt.title('Map')
plt.show()
```

The rider almost completes the lap. So the exact point where he completes the lap is around 378 seconds


```python
# Map for first 375 data points
fig = plt.figure(figsize=(10, 10))
plt.scatter(dataset['Longtitude'][0:378], dataset['Latitude'][0:378])
plt.title('Map')
plt.show()
```

So after 378 seconds, he completed one lap, we therefore can calculate how far each lap is:

```python
distance_in_meters = []
for i in range(0, 378):
    starting_point = (dataset['Longtitude'][i], dataset['Latitude'][i])
    ending_point = (dataset['Longtitude'][i+1], dataset['Latitude'][i+1])
    distance = vincenty(starting_point, ending_point).meters
    distance_in_meters.append(distance)
total_distance_onelap = sum(distance_in_meters)
print(total_distance_onelap)
```

We will round up to 3500 meters each lap. Then, divide total_distance for total_distance_onelap, we can have the number of lap rider 1 has completed is 18 laps


```python
num_laps_rider1 = total_distance/ total_distance_onelap
print(round(num_laps_rider1,3))
```

```python
# Similarly, read data in to a dataframe
tree = ET.parse('race2.gpx')
root = tree.getroot()

# define a namespace dictionary to make element names simpler
# this mirrors the namespace definitions in the XML files
ns ={'gpx': 'http://www.topografix.com/GPX/1/1', 'gpxtpx': 'http://www.garmin.com/xmlschemas/TrackPointExtension/v1'}

# when we look for elements, we need to use the namespace prefix
trk = root.find('gpx:trk', ns)
trkseg = trk.find('gpx:trkseg', ns)

# iterate all trkpt elements - the children of trkseg
data = []
for trkpt in trkseg:
    variables = []
    # get some properties from the attributesd
    lon= trkpt.attrib['lon']
    lat = trkpt.attrib['lat']
    # get values from the child elements
    ele = trkpt.find('gpx:ele', ns).text
    variables = [lon, lat, ele]
    data.append(variables)
    
# read data into a dataframe and name the variables
columns = ['Longtitude', 'Latitude', 'Elevation']
dataset2 = pd.DataFrame(data, columns = columns)

# convert three variables to float type
for i in ['Longtitude', 'Latitude', 'Elevation']:
    dataset2[i] = dataset2[i].astype(float)
```

```python
# Quick visualization of the map the rider 2 passed through
fig = plt.figure(figsize=(10, 10))
plt.scatter(dataset2['Longtitude'][0:(len(dataset2)-1)], dataset2['Latitude'][0:(len(dataset2)-1)])
plt.title('Map 2')
plt.show()
```

```python
distance_in_meters = []
for i in range(0, len(dataset2) - 1):
    starting_point = (dataset2['Longtitude'][i], dataset2['Latitude'][i])
    ending_point = (dataset2['Longtitude'][i+1], dataset2['Latitude'][i+1])
    distance = vincenty(starting_point, ending_point).meters
    distance_in_meters.append(distance)
total_distance_2 = sum(distance_in_meters)
print('Total distance of rider 2 is', round(total_distance_2,2), 'm')

# Average speed of the rider 2 in m/s
average_speed_ms_2 = round(total_distance_2 / (len(dataset2)), 2)
print("The average speed of the rider 2 is", average_speed_ms_2, "m/s")

# Average speed of the rider 2 in m/s
average_speed_kmh_2 = round(average_speed_ms * 18/5, 2)
print("The average speed of the rider 2 is",average_speed_kmh_2, "km/h")
```

Assumed the reading is taken every one second, we apply a similar method and calculate the average speed of rider 2, which is 10.11 m/s in this case. This is probably a true assumption because the speed of rider 2 is comparable to rider 1's.
One more thing, the rider 2 did cycle on a nearly identical lap as we can see in the above visualization. 


```python
speed_diff = average_speed_ms_2 - average_speed_ms
print("Rider 2 is faster than rider 1:", round(speed_diff,2), "m/s in average")
```

```python
#Time take for each lap of rider 1:
num_laps_rider2 = total_distance_2/ total_distance_onelap
print(round(num_laps_rider2,3))
```

Rider 2 has finished nearly 21 laps. Base on the total time cycled of two riders, we can easily calculate the time for each lap:

```python
#Rider 1:
time_for_each_lap_1 = len(dataset)/num_laps_rider1
print('Time taken for each lap by rider 1 is', round(time_for_each_lap_1,2), 'seconds')

#Rider 2:
time_for_each_lap_2 = len(dataset2)/num_laps_rider2
print('Time taken for each lap by rider 2 is', round(time_for_each_lap_2,2), 'seconds')
```
