# Online Retail Customer Clustering

This exercise uses a dataset containing records of online retail transactions. The data is available at [The UCI ML Repository](https://archive.ics.uci.edu/ml/datasets/Online+Retail).  

The goal of the exercise is to use clustering techniques to idenfity customer segments in the data that might be suitable for targetted marketing or market research. The data is available as a spreadsheet, I've saved this as a CSV file.   I've written some code below to read 
the data and to add some derived columns to the data frame.  I've done this, rather than ask you to do it, 
because it was fairly tricky to find out how to do this and uses groupby and aggregate functions that we've not
covered much.  Take this as an example of data preparation.


```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
import matplotlib.pylab as plt
import seaborn as sb
%matplotlib inline

retail = pd.read_csv('files/online-retail.csv')
retail.head()
```

```python
# Boxplots to detect outliers
fig_1 = sb.boxplot(retail['UnitPrice'], orient='v')
fig_1.set_title('Boxplot for Unit Price')
plt.show()
fig_2 = sb.boxplot(retail['Quantity'], orient='v')
fig_2.set_title('Boxplot for Quantity')
plt.show()
```

We should remove any data points with negative Unit Price - which is not appropriate. We will only keep data points with UnitPrice less than 20,000 <br> In addition, we should also remove outliers with large Quantity (greater than 60,000 or less than -60,000).

```python
# Remove outliers 
retail = retail[retail.UnitPrice >= 0]
retail = retail[retail.UnitPrice < 20000]
retail = retail[retail.Quantity < 20000]
retail = retail[retail.Quantity > -20000]
```

```python
# add a column 'value' for the value of each transaction
retail = retail.assign(Value=retail.Quantity * retail.UnitPrice)

# split into returns and sales, returns have a quantity less than zero
returns = retail.loc[retail.Quantity<=0]
sales = retail.loc[retail.Quantity>0]
print("Sales", sales.shape)
print("Returns", returns.shape)
```

## Adding Columns

This next step turns the data into a dataframe with customers as the index and aggregates the numerical columns using the mean.  So we get the mean quantity, price and value of transactions for this customer.  I also add a count of the number of transactions per customer. 

You should experiment with the groupby and aggregate methods, they're very useful.

```python
sales_grouped = sales[['CustomerID', 'Quantity', 'UnitPrice', 'Value']].groupby('CustomerID').aggregate(np.mean)

# add a count of the number of invoices (== frequency in referenced analysis)
counts = sales.groupby('CustomerID').size().to_frame(name='Count')
sales_grouped = sales_grouped.join(counts)
sales_grouped.head()
```

This next step computes the time from each invoice date to the most recent invoice date and converts it to days. The following step then aggregates this per customer using *max* to get the most recent invoice per customer.  This is then added to the *sales_grouped* dataframe. 

```python
# add recency of purchases
# this takes some time to compute
recency = pd.to_datetime(sales.InvoiceDate) - pd.to_datetime("2011-12-10")
# get just the number of days so that we can use this value in clustering
recency = recency.apply(lambda x: x.days)
```

```python
# turn recency into a dataframe with the customerID column from sales
recency_df = sales[['CustomerID']].assign(recency=recency)
# aggregate recency using max to get the most recent sale per customer
grp = recency_df[['CustomerID', 'recency']].groupby('CustomerID').aggregate(np.max)

# customers is our final dataframe of customer data 
customers = sales_grouped.join(grp)
# exclude customers with Value equals zero
customers = customers[customers['Value'] != 0]
customers.head()
```

## Your Analysis


The main purpose of this clustering is discovering high and low value customer segments for marketing purposes. <br>

For simplicity and more efficient computation, K-means clustering algorithm will be used in the analysis.
In case of such a large dataset (4338 rows), K-means will return the results much faster than other clustering algorithms (e.g. Hirerachical) due to not having to compute distance matrix for each pair of data points. <br>

The number of pre-defined clusters (k) will be trials and errors, whatever sensible and explainable number of k will be accepted (arbitrary). However, let's make a scatter plot to get an holistic view of relationships among Value, Count and Recency. <br>


```python
#Scatter plot for Frequency and Value, with recency as another categorizer.
plt.scatter(customers['Count'], customers['Value'], c=customers['recency'], cmap=plt.cm.coolwarm)
clb = plt.colorbar()
clb.ax.set_title('Recency')
plt.xlabel('Frequency')
plt.ylabel('Value')
plt.title('Scatter plot for Frequency and Value')
plt.show()
```

The graph is not interpretable because almost the data points gathering in the bottom left of the graph. We need to do some Log transformation and standardization for all three variables.



```python
# Log Transformation for Count and Value
customers['Count_Log'] = np.log(customers['Count'])
customers['Value_Log'] = np.log(customers['Value'])
# Standardize recency variable
customers['Recency_Std'] = (customers['recency'] - np.mean(customers['recency'])) / np.std(customers['recency'])
```

```python
#Scatter plot for Log of Frequency and Log of Value, with recency as another categorizer.
plt.scatter(customers['Count_Log'], customers['Value_Log'], c=customers['Recency_Std'], cmap=plt.cm.coolwarm)
clb = plt.colorbar()
clb.ax.set_title('Recency')
plt.xlabel('Log Frequency')
plt.ylabel('Log Value')
plt.title('Scatter plot for Log Frequency and Log Value')
plt.show()
```

The warmer the color, the more recent a customer purchases an order. <br>
The less warm the color, the less recent a cusomer purchases an order.<br>
As we can see from the graph, there are no obvious boundaries among group of customers. Most recent and most frequent customers usually buy products with low to medium prices. A little less frequent and less recent customers purchase low to medium, and sometimes high valued products. A couples of low-frequent customers purchase really high valued products sometimes. <br> 

We should proceed to K-means clustering for more obvious results. We will experiment with a range of number of clusters and hopefully we can define clearer boundaries among clusters.  <br>




```python
# Define K-means clustering model for k clusters

def k_means_alg(k):
    clf = KMeans(init='k-means++', n_clusters=k)
    k_features = ['Count_Log', 'Value_Log', 'Recency_Std']
    clf.fit(customers[k_features])
    labels = clf.predict(customers[k_features])
    customers['Cluster'+str(k)] = labels
#Scatter plot for Log of Frequency and Log of Value, with cluster as a categorizer.
    colour = ['r', 'g', 'b', 'y', 'black']
    for i in range(k):
        plt.scatter(customers['Count_Log'][customers['Cluster'+str(k)]==i], customers['Value_Log'][customers['Cluster'+str(k)]==i], 
                    c=colour[i], label='Cluster '+str(i))

    plt.legend()
    plt.xlabel('Log Frequency')
    plt.ylabel('Log Value')
    plt.title('k-Means with ' + str(k) + ' clusters')
    plt.show()

#Summary table in orginal scale
    summary_table_org = pd.DataFrame(clf.cluster_centers_)
    summary_table_org.columns = ['Frequency','Value', 'Recency']
    summary_table_org.index.name = 'Cluster' 
    summary_table = np.exp(summary_table_org[['Frequency','Value']])
    summary_table['Recency'] = (summary_table_org[['Recency']] * np.std(customers['recency'])) + np.mean(customers['recency'])
    print(summary_table)
    
```

We will use the above function to visualize clusters in several cases. After that, we can base on the summary table, indicated Frequency, Value and Recency of each cluster centroid in orinigal scale, to explain that nature of each cluster. As a result, we  can provide reasonable explanations and recommendations to the marketing team. <br>

Firstly, let make a summary statistics table for comparision purpose.


```python
customers[['Value', 'Count', 'recency']].describe()
```

```python
# 2 clusters solution
result = k_means_alg(2)
```

The 2 clusters solution provides: <br>
_ Cluster 0: medium frequent, medium value, high recent. <br>
_ Cluster 1: low frequent, medium value, low recent. <br>
It is not clear which customer segment we should focus on since the median Values of two cluster centroids are lower than the average Value as a whole. 

```python
# 3 clusters solution
k_means_alg(3)
```

The 3 clusters solution provides: <br>
_ Cluster 0: medium frequent, low value, high recent. <br>
_ Cluster 1: very low frequent, low value, very low recent. <br>
Cluster 2: very low frequent, medium value, medium recent.<br>
Again, we cannot distinguish obvious boundaries among clusters. However, if this model is used, we at least have a segment to avoid - Cluster 1 - which contains customers who rarely visit the store, buy less valued products and have not visited the store for a long time. 

```python
# 4 clusters solution
k_means_alg(4)
```

The 4 clusters solution provides: <br>
_ Cluster 0: high frequent, low value, high recent. <br>
_ Cluster 1: very low frequent, low value, high recent. <br>
_ Cluster 2: very low frequent, really high value, low recent. <br>
_ Cluster 3: very low frequent, low value, very low recent.<br>
For this model, we should focus on marketing to Cluster 0 - customers who visit regularly and recently. Another Cluster to pay attention is number 2 - they bought really high valued products but do not visit usually and recently. 

```python
# 5 clusters solution
k_means_alg(5)
```

The 5 clusters solution provides: <br>
_ Cluster 0: high frequent, medium value, high recent. <br>
_ Cluster 1: very low frequent, really high value, low recent. <br>
_ Cluster 2: very low frequent, low value, really low recent.<br>
_ Cluster 3: low frequent, medium value, high recent.<br>
_ Cluster 4: very high frequent, really low value, really high recent. <br>
This model clearly explains the segments: <br>
_ Very loyal customers (cluster 4), come usually and recently, but buy daily products - which are low in values (makes sense). Therefore, we should not focus on marketing to this group (they will come anyway). Instead, implementing loyalty program to keep them with the business (4th priority) <br>
_ Loyal customers (cluster 0) who also come usually and recently, but buy more valued products than the first group - therefore the most priority group - need to be paid attention on. <br>
_ Unusual but valued customers (cluster 1) who do not come usually and recently, but buy really high valued products. This group should be the second priority to marketing to in order to promote more usual purchases.<br>
_ New customers (cluster 3) who come quite recently, but not frequently and buy medium valued products. This is the third priority.<br>
_ Zero-value customers (cluster 2) - have not visited the store for a long time , have not visited the store many times and have not bought valued products. No marketing to this group.

As we move beyond 5 clusters, the segments become less distinguished and less explainable. Therefore, our final recommendation is to employ the 5 clusters model for marketing and promotion strategies.
